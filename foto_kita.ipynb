```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# Menyimpan riwayat pengecekan BMI
riwayat = []

def klasifikasi_bmi(bmi):
    if bmi < 18.5:
        return "Kurus", "#F0A500"
    elif bmi < 25:
        return "Normal", "#2E8B57"
    elif bmi < 30:
        return "Gemuk", "#E08300"
    else:
        return "Obesitas", "#D04444"

# Input widget
input_berat = widgets.FloatText(description="Berat (kg):", value=None, style={"description_width": "100px"})
input_tinggi = widgets.FloatText(description="Tinggi (cm):", value=None, style={"description_width": "100px"})
tombol_hitung = widgets.Button(description="Hitung BMI", button_style="primary", icon="calculator")
tombol_reset = widgets.Button(description="Hapus riwayat", button_style="", icon="trash")
output_hasil = widgets.Output()
output_grafik = widgets.Output()

def gambar_grafik():
    with output_grafik:
        clear_output(wait=True)
        if not riwayat:
            print("Belum ada data riwayat. Hitung BMI pertamamu dulu.")
            return

        nomor_cek = list(range(1, len(riwayat) + 1))
        nilai_bmi = [r["bmi"] for r in riwayat]

        plt.figure(figsize=(7, 4))
        plt.plot(nomor_cek, nilai_bmi, marker="o", color="#378ADD", linewidth=2)
        plt.fill_between(nomor_cek, nilai_bmi, alpha=0.1, color="#378ADD")

        # Garis batas kategori BMI
        plt.axhline(18.5, color="#F0A500", linestyle="--", linewidth=0.8, label="Batas kurus/normal")
        plt.axhline(25, color="#E08300", linestyle="--", linewidth=0.8, label="Batas normal/gemuk")
        plt.axhline(30, color="#D04444", linestyle="--", linewidth=0.8, label="Batas gemuk/obesitas")

        plt.xticks(nomor_cek, [f"Cek {i}" for i in nomor_cek])
        plt.ylabel("Nilai BMI")
        plt.title("Riwayat Pengecekan BMI")
        plt.legend(fontsize=8, loc="upper left")
        plt.ylim(12, 40)
        plt.grid(axis="y", linestyle=":", alpha=0.5)
        plt.tight_layout()
        plt.show()

        print("\nTabel riwayat:")
        print(f"{'No':<4}{'Berat':<10}{'Tinggi':<10}{'BMI':<8}{'Kategori':<10}")
        for i, r in enumerate(riwayat, start=1):
            print(f"{i:<4}{r['berat']:<10}{r['tinggi']:<10}{r['bmi']:<8}{r['kategori']:<10}")

def saat_hitung_diklik(b):
    with output_hasil:
        clear_output(wait=True)
        berat = input_berat.value
        tinggi = input_tinggi.value

        if not berat or not tinggi or berat <= 0 or tinggi <= 0:
            print("Isi berat dan tinggi badan dengan angka yang valid.")
            return

        tinggi_m = tinggi / 100
        bmi = berat / (tinggi_m ** 2)
        bmi = round(bmi, 1)
        kategori, warna = klasifikasi_bmi(bmi)

        riwayat.append({"berat": berat, "tinggi": tinggi, "bmi": bmi, "kategori": kategori})

        print(f"BMI kamu: {bmi}  ->  Kategori: {kategori}")

    gambar_grafik()

def saat_reset_diklik(b):
    riwayat.clear()
    with output_hasil:
        clear_output(wait=True)
        print("Riwayat dihapus.")
    gambar_grafik()

tombol_hitung.on_click(saat_hitung_diklik)
tombol_reset.on_click(saat_reset_diklik)

display(widgets.HBox([input_berat, input_tinggi]))
display(widgets.HBox([tombol_hitung, tombol_reset]))
display(output_hasil)
display(output_grafik)

gambar_grafik()
```
